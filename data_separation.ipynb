### Separate dataset into training and test sets

```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
```

```python
song_df = pd.read_csv('data/hannah_lastfm.csv')

# remove songs w/o track ids
song_df.dropna(subset=['track_mbid'], inplace=True)

# add listen count and remove duplicate songs
vals = song_df.value_counts('track_mbid')
song_df = song_df.join(vals, on='track_mbid')
song_df.drop_duplicates(subset='track_mbid', inplace=True)

print(song_df.info)
```

```python
SEED = 64
train_set, test_set = train_test_split(song_df, test_size=0.2, random_state=SEED)

print(train_set.info())
print(test_set.info())
```
