### Separate dataset into training and test sets

```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
```

```python
SEED = 64

song_df = pd.read_csv('data/hannah_lastfm.csv')
train_set, test_set = train_test_split(song_df, test_size=0.2, random_state=SEED)
```
